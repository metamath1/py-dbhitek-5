````python
# ============================================================
# 웨이퍼 개수와 불량칩의 개수를 입력하면
# 수율을 계산하는 프로그램
# ============================================================
# 이 예제는 실제 생산 시스템에서 데이터를 가져오는 것이 아닙니다.
# 웨이퍼 개수, 웨이퍼당 칩 개수, 불량 칩 개수가
# 1초마다 들어오는 상황을 단순화한 교육용 예제입니다.
#
# 수율 계산식 역시 문법 학습을 위해 단순화한 계산식입니다.
# 실제 현업에서는 공정, 제품, 검사 기준, 재작업 여부 등에 따라
# 수율 정의가 달라질 수 있습니다.
#
# 학습 목표:
# - 함수를 정의한다.
# - 매개변수를 사용한다.
# - 계산 결과를 return 한다.
# - 반복적으로 들어오는 데이터에 같은 함수를 재사용한다.
#
# 수강생은 calculate_yield() 함수 내부만 작성합니다.
# random과 time은 실시간 데이터처럼 보이게 하기 위한 제공 코드입니다.
# ============================================================

import random
import time


def calculate_yield(wafer_count, chips_per_wafer, defective_chips):
    # ========================================================
    # 수강생 작성 영역
    #
    # 목표:
    # 웨이퍼 개수, 웨이퍼 1장당 칩 개수, 불량 칩 개수를 사용해
    # 수율을 계산하고 return 하세요.
    #
    # 교육용 단순 계산식:
    # 전체 칩 수 = 웨이퍼 개수 * 웨이퍼 1장당 칩 개수
    # 정상 칩 수 = 전체 칩 수 - 불량 칩 수
    # 수율 = 정상 칩 수 / 전체 칩 수 * 100
    #
    # 주의:
    # 이 계산식은 파이썬 함수 학습을 위한 단순화된 수율 계산입니다.
    # 실제 제조 현장의 수율 계산은 제품/공정/검사 기준에 따라
    # 더 복잡할 수 있습니다.
    # ========================================================

    ###########################################################################
    # 여기에 코드 작성



    ###########################################################################

    # 아래 return 0은 코드가 실행되도록 넣어둔 임시값입니다.
    # 과제 수행 시 실제 계산 결과를 return 하도록 수정하세요.
    return 0


print("[실시간 웨이퍼 수율 계산 프로그램]")
print("1초마다 생산 데이터가 들어온다고 가정합니다.")
print("아래 데이터는 파이썬 문법 학습을 위한 교육용 가상 데이터입니다.\n")


for i in range(10):
    # --------------------------------------------------------
    # 교육용 가상 생산 데이터 생성
    # 실제 생산 설비에서 가져온 데이터가 아닙니다.
    # --------------------------------------------------------
    wafer_count = random.randint(1, 5)
    chips_per_wafer = random.randint(500, 800)

    total_chips = wafer_count * chips_per_wafer

    # 전체 칩 수의 최대 약 8% 범위 안에서 불량 칩 개수를 임의 생성합니다.
    # 이 값 역시 실제 불량률을 의미하지 않는 교육용 가상값입니다.
    defective_chips = random.randint(0, int(total_chips * 0.08))

    yield_rate = calculate_yield(
        wafer_count,
        chips_per_wafer,
        defective_chips
    )

    print("측정 번호:", i + 1)
    print("웨이퍼 개수:", wafer_count)
    print("웨이퍼 1장당 칩 개수:", chips_per_wafer)
    print("불량 칩 개수:", defective_chips)
    print("수율:", round(yield_rate, 2), "%")
    print("-" * 50)

    time.sleep(1)
````
